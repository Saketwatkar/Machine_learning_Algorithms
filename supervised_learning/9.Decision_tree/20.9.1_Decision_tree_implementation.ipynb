```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris['DESCR']
```

```python
iris['target']
```

```python
x=pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
x ## independent variables or features
```

```python
y=iris['target'] ## dependent variable or target
```

```python
y
```

```python
# train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
## Apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(max_depth=2)
```

```python
dtree.fit(x_train, y_train)
```

```python
## visualize the decision tree
from sklearn import tree
plt.figure(figsize=(8,7))
tree.plot_tree(dtree, filled=True)
plt.show()
```

```python
dtree.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
y_pred = dtree.predict(x_test)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

## Decision tree classifier prepruning and Hyperparameter tuning

```python
param={
    'criterion':['gini', 'entropy'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
dtree = DecisionTreeClassifier()
grid = GridSearchCV(dtree, param_grid=param, cv=5, scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings("ignore")
# Fit the model with the training data
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
score= accuracy_score(y_test, y_pred)
print("Accuracy Score:", score)
```
